## Experimental Notes

The lifetime of an atom $\tau$ is extracted from the phase lag $\phi$ via the relation:

$$
\begin{align}
    \tan(\phi) = \omega \tau \Rightarrow \tau = \frac{\tan(\phi)}{2\pi f}
\end{align}
$$

where $\omega = 2\pi f$ and $f$ is the input modulation frequency.

## What to record from Moku:lab


\begin{aligned}
&\bullet \text{Modulation frequency $f$ [Hz]} \\
&\bullet \text{Lock-in outputs X and Y [Voltz]} \\
&\bullet \text{}
\end{aligned}

\begin{aligned}
&\bullet \text{$\tan(\phi) = \frac{Y}{X}$} \\
&\bullet \text{$\tau = \frac{\tan(\phi)}{\omega}$} \\
&\bullet \text{Can take multiple trials, average and take SEM to do experimental analysis and model fitting for confidence estimates}
\end{aligned}
